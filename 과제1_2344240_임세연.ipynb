Car 클래스는 속도(speed)와 무게(weight) 필드를 가지고 있고, 메소드로는 속도를 올리는 speedUp, 속도를 내리는 speedDown이 있으며, 무게를 무겁게하는 weightUp이 있다.  
이때, ModelA 클래스는 최대 속도가 250km/h, 최대 무게가 2,500kg으로 제한되도록 ModelA 클래스를 만드시오.

````python
class Car:
  count = 0
  def __init__(self, speed=0, weight=0):
    self.count += 1
    self.speed = float(speed)
    self.weight = float(weight)

  def speedUp(self):
    self.speed += 10

  def speedDown(self):
    self.speed -= 10

  def weightUp(self):
    self.weight += 100

  def printAll(self):
    print(f'전체 차량 수: {self.count:>10}')
    print(f'현재 차량의 속도: {self.speed:>10}km/h')
    print(f'현재 차량의 무게: {self.weight:>10}kg')
````

````python
class Car:
    count = 0
    def __init__(self, speed=0, weight=0):
        Car.count+=1
        self.speed = float(speed)
        self.weight = float(weight)

    def speedUp(self):
        self.speed += 10

    def speedDown(self):
        self.speed -= 10
        if self.speed < 0:
            self.speed = 0  # 음수 속도가 되지 않도록 설정

    def weightUp(self):
        self.weight += 100

    def printAll(self):
        print(f'전체 차량 수: {Car.count:>10}')
        print(f'현재 차량의 속도: {self.speed:>10}km/h')
        print(f'현재 차량의 무게: {self.weight:>10}kg')

class ModelA(Car):
    max_speed = 250              # 최대속도, 최대무게 지정
    max_weight = 2500

    def __init__(self, speed=0, weight=0):
        super().__init__(speed, weight)          # super는 부모클래스를 가져다 쓰는 거임
                                                 # 위에 있는 init 그대로 들고온다
    def speedUp(self):
        if self.speed < self.max_speed:
            super().speedUp()                  # Car클래스에 speedUp 함수 호출
        else:
            print("최대 속도를 초과하였습니다 더 이상 속도를 늘릴 수 없습니다 주의하세요.")

    def speedDown(self):
            super().speedDown()                # Car클래스에 speedDown 함수 호출

    def weightUp(self):
        if self.weight  < self.max_weight:
            super().weightUp()                 # Car클래스에 weightUp 함수 호출
        else:
            print("최대 무게를 초과하였습니다 더 이상 무게를 늘릴 수 없습니다 주의하세요.")
````

(단, 파일 경로는 bubbleSort.py이며, 함수명은 bubbleSort이다.)  
이때, time 모듈을 import하여 시간을 측정한다.  

time 모듈은 아래와 같이 사용한다.  
```python

````python
%%writefile bubbleSort.py
def bubbleSort(arr):
    n = len(arr)

    # Traverse through all array elements
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):

            # Traverse the array from 0 to n-i-1. Swap if the element found
            # is greater than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
````

````python
import time
from bubbleSort import bubbleSort

# 임의의 값 10개 생성
arr = [3, 7, 2, 5, 9, 4, 6, 1, 8, 10]

# 시간 측정 시작
start = time.time()

# bubbleSort 함수 호출
bubbleSort(arr)

# 시간 측정 종료 및 소요된 시간 출력
end = time.time() - start

print("정렬된 배열:", arr)
print("소요된 시간:", end, "초")
````
