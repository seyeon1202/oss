1. 재귀 초급 문제  
팰린드롬이란, 앞에서부터 읽었을 때와 뒤에서부터 읽었을 때가 같은 문자열을 말한다.  
어떤 문자열이 팰린드롬인지 판별하는 문제는 재귀 함수를 이용해 쉽게 해결할 수 있다. 아래 코드의 isPalindrome 함수는 주어진 문자열이 팰린드롬이면 1, 팰린드롬이 아니면 0을 반환하는 함수다.  
동민이는 위에 작성된 isPalindrome 함수를 이용하여 어떤 문자열이 팰린드롬인지 여부를 판단하려고 한다.  
문자열 S를 isPalindrome 함수의 인자로 전달하여 팰린드롬 반환값과 판별하는 과정에서 recursion 함수를 몇 번 호출하는지 아래와 같이 출력하는 프로그램을 코딩하시오.


```python
def recursion(s, l, r):
    if l >= r: return 1
    elif s[l] != s[r]: return 0
    else: return recursion(s, l+1, r-1)  

def isPalindrome(s):
    return recursion(s, 0, len(s)-1)

print('ABBA:', isPalindrome('ABBA'))
print('ABC:', isPalindrome('ABC'))
```

입력

```
5
AAA
ABBA
ABABA
ABCA
PALINDROME
```
출력

```
1 2
1 3
1 3
0 2
0 1
```



````python
def recursion(s, l, r, count):
    if l >= r:          # l이 r보다 크거나 같으면, 문자열의 중간 지점을 넘어섰으므로 팰린드롬임을 의미합니다.
        return 1, count
    elif s[l] != s[r]:  # 문자열 s의 왼쪽 끝과 오른쪽 끝의 문자가 다르면, 팰린드롬이 아닙니다.
        return 0, count
    else:
        count += 1      # count를 증가시킵니다.
        return recursion(s, l + 1, r - 1, count)  # 다음 문자열 쌍을 확인하기 위해 재귀 호출합니다.

def is_palindrome(s):
    count = 1       # 팰린드롬 확인 횟수를 1로 초기화합니다.
    result, count = recursion(s, 0, len(s) - 1, count)  # recursion 함수를 실행하여 팰린드롬 여부를 확인합니다.
    return result, count

# 입력
n = int(input())  # 문자열의 개수를 입력받습니다.
s = []            # 문자열을 저장할 리스트를 생성합니다.
for _ in range(n):
    x = input()   # n개의 문자열을 입력받습니다.
    s.append(x)   # 입력받은 문자열을 리스트에 저장합니다.
for s1 in s:
    result, count = is_palindrome(s1) # 각 문자열에 대해 팰린드롬 여부를 확인합니다.
    print(result, count) # 결과를 출력합니다.
````
