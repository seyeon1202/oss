```python
import pygame

# 초기 설정
pygame.init()
screen_width = 800
screen_height = 600
screen = pygame.display.set_mode((screen_width, screen_height))
clock = pygame.time.Clock()

# 게임 에서 사용할 색상 정의(RGB 코드)
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)
GREEN = (0, 255, 0)
BLUE = (0, 0, 255)

# 막대기(패들) 객체 정의
paddle_width = 100
paddle_height = 12
paddle_speed = 7
paddle = pygame.Rect(screen_width//2-paddle_width//2,
                     screen_height-40,
                     paddle_width,
                     paddle_height)

# 벽돌 객체 정의
brick_width = 60
brick_height = 20
bricks = []

# 공 정의
ball_diameter = 12
ball_speed = [5, -5]
ball = pygame.Rect(screen_width//2 - ball_diameter//2,
                   screen_height//2-ball_diameter//2,
                   ball_diameter,
                   ball_diameter)

# 가로10개, 세로5개의 벽돌을 생성
for x in range(10):
    for y in range(5):
        brick = pygame.Rect(x*(brick_width+10)+35,
                            y*(brick_height+10)+35,
                            brick_width,
                            brick_height)
        bricks.append(brick)

# 점수 설정
score = 0

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # 막대기 (패들)
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and paddle.left:
        paddle.left -= paddle_speed
    if keys[pygame.K_RIGHT] and paddle.right < screen_width:
        paddle.right += paddle_speed

    # 공 움직임
    ball.left += ball_speed[0]
    ball.top += ball_speed[1]

    # 공과 벽의 충돌 감지
    if ball.left <= 0 or ball.right >= screen_width:
        ball_speed[0] = -ball_speed[0]
    if ball.top <= 0:
        ball_speed[1] = -ball_speed[1]
    if ball.colliderect(paddle):
        ball_speed[1] = -ball_speed[1]

    # 공과 벽돌의 충돌
    for brick in bricks[:]:
        if ball.colliderect(brick):
            if ball_speed[0] > 0 and brick.left + 1 == ball.right:
                ball_speed[0] = -ball_speed[0]
            elif ball_speed[0] < 0 and brick.right == ball.left + 1:
                ball_speed[0] = -ball_speed[0]
            elif brick.left < ball.right and ball.left < brick.right:
                ball_speed[1] = -ball_speed[1]

            bricks.remove(brick)
            score += 10  # 벽돌을 깰 때마다 10점씩 증가

    # 배경색 그리기
    screen.fill(BLACK)

    # 막대기, 공, 벽돌 그리기
    pygame.draw.rect(screen, GREEN, paddle)
    pygame.draw.ellipse(screen, WHITE, ball)
    for brick in bricks:
        pygame.draw.rect(screen, RED, brick)

    # 점수 및 승리/패배 표시
    font = pygame.font.Font(None, 36)
    score_text = font.render(f"Score: {score}", True, WHITE)
    screen.blit(score_text, (10, 10))

    if not bricks:
        game_over_text = font.render("win!", True, WHITE)
        screen.blit(game_over_text, (screen_width//2 - 50, screen_height//2))
        pygame.display.flip()
        pygame.time.delay(1000)  # 1초 대기
        running = False

    if ball.bottom >= screen_height:
        game_over_text = font.render("Game over", True, WHITE)
        screen.blit(game_over_text, (screen_width//2 - 50, screen_height//2))
        pygame.display.flip()
        pygame.time.delay(1000)  # 1초 대기
        running = False

    pygame.display.flip()
    clock.tick(60)

print("score:", score)
```
