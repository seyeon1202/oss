```python
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board, player):
    """승자를 체크하는 함수"""
    # 가로, 세로, 대각선 체크
    for i in range(3):
        if all(board[i][j] == player for j in range(3)):  # 가로 체크
            return True
        if all(board[j][i] == player for j in range(3)):  # 세로 체크
            return True
    if all(board[i][i] == player for i in range(3)):  # 왼쪽 위에서 오른쪽 아래 대각선 체크
        return True
    if all(board[i][2-i] == player for i in range(3)):  # 오른쪽 위에서 왼쪽 아래 대각선 체크
        return True
    return False

def is_board_full(board):
    """보드가 가득 찼는지 확인하는 함수"""
    for row in board:
        if " " in row:
            return False
    return True

def player_move(board):
    """사용자의 움직임을 처리하는 함수"""
    while True:
        try:
            row = int(input("행을 선택하세요 (1-3): ")) - 1
            col = int(input("열을 선택하세요 (1-3): ")) - 1
            if board[row][col] == " ":
                board[row][col] = "X"
                break
            else:
                print("이미 선택된 위치입니다. 다시 선택해주세요.")
        except (ValueError, IndexError):
            print("잘못된 입력입니다. 다시 입력해주세요.")

def computer_move(board):
    """컴퓨터의 움직임을 처리하는 함수"""
    while True:
        row = random.randint(0, 2)
        col = random.randint(0, 2)
        if board[row][col] == " ":
            board[row][col] = "O"
            break

def tic_tac_toe_game():
    """틱택토 게임을 실행하는 함수"""
    board = [[" " for _ in range(3)] for _ in range(3)]
    current_player = "X"  # 사용자가 먼저 시작

    while True:
        print_board(board)

        if current_player == "X":
            player_move(board)
        else:
            computer_move(board)

        # 승자 체크
        if check_winner(board, current_player):
            print_board(board)
            print(f"{current_player}가 이겼습니다!")
            break

        # 비김 체크
        if is_board_full(board):
            print_board(board)
            print("무승부입니다!")
            break

        # 플레이어 변경
        current_player = "O" if current_player == "X" else "X"

# 게임 실행
if __name__ == "__main__":
    tic_tac_toe_game()
```
